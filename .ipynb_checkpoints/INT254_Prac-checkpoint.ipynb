```python
import pandas as pd
```

```python
data =pd.read_csv("Data.csv")
print(data)
```

```python
#using .get_dummies functions to convert
#the categorial datatypes to numerical and storing the returned dataframes in a new variables
data1=pd.get_dummies(data['Purchased'])
data=pd.concat([data, data1], axis=1)
data.drop('Purchased', axis=1, inplace=True)
data
```
